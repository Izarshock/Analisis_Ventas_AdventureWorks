```python
import pandas as pd
import pyodbc
```

```python
server = 'localhost\\SQLEXPRESS'
database = 'AdventureWorks2017'
data_string = (
    f'DRIVER={{ODBC Driver 17 for SQL Server}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    f'Trusted_Connection=yes;'
)
string = pyodbc.connect(data_string)

query = """
SELECT
    T2.Name As SalesTerritory,
    T1.TotalDue As TotalRevenue
From Sales.SalesOrderHeader As T1
Inner Join Sales.SalesTerritory As T2
    On T1.TerritoryID = T2.TerritoryID
"""

df_ventas = pd.read_sql(query, string)

print('DataFrame cargado de la base de datos real:')
print(df_ventas.head(50))
print(df_ventas.info())
```

```python
df_reporte = df_ventas.groupby('SalesTerritory')['TotalRevenue'].sum()
print(df_reporte)
```
